```python
import pandas as pd
import numpy as np
import os
from bokeh.plotting import figure, output_file, show
from bokeh.models.ranges import Range1d
#from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20, Inferno
from bokeh.models import HoverTool, Plot
from bokeh.models.glyphs import MultiLine
```

```python
path = os.getcwd()
file0 = os.path.join(path, "econ-rates.csv")
#outfile = os.path.join(path, "Lease Excelwriter.xlsx")
dg = pd.read_csv(file0)

dgh = pd.DataFrame(dg,   columns = ['Desc', 'date', 'OBS_VALUE'])
#dg['date2'] = pd.to_datetime(dg['date'])

base = dgh[(dgh['OBS_VALUE'] >=-50) & (dgh['OBS_VALUE'] <=50) ]
```

```python
grp_list = base['Desc'].unique()
```

```python
xs = [base.loc[base.Desc == i]['date'].values for i in grp_list]
ys = [base.loc[base.Desc == i].OBS_VALUE.values for i in grp_list]


source = ColumnDataSource(data=dict(
     x = xs,
     y = ys, 
    color = (Category20[7])[0:len(grp_list)],
     group = grp_list))

p3 = figure(plot_width=900, plot_height=300)

p3.multi_line(
     xs='x',
     ys='y',
     legend='group',
     source=source,
     line_color='color')
```
